```python
x = np.array([100, 200, 300, 400, 500])
y = np.array([10, 20, 30, 40, 50])
plt.plot(
    x, y,
    x, np.power(y, 1.1),
    x, np.power(y, 1.2),
    )
plt.show()
```

```python
x = np.array([100, 200, 300, 400, 500])
y = np.array([10, 20, 30, 40, 50])
plt.plot(
    x, y, 'b-',
    x, np.power(y, 1.1), 'r-o',
    x, np.power(y, 1.2), 'g*:',
    )
plt.show()
```

### figsize

### title

```python
plt.figure(figsize=(10, 5))
plt.plot([100, 200, 300, 400, 500], [10, 20, 30, 40, 50])

plt.title("line chart")
plt.show()
```

### xlabel, ylabel

```python
plt.figure(figsize=(10, 5))
plt.plot([100, 200, 300, 400, 500], [10, 20, 30, 40, 50], label="blue")

plt.title("line chart")
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```

### grid

```python
plt.figure(figsize=(10, 5))
plt.plot([100, 200, 300, 400, 500], [10, 20, 30, 40, 50], label="blue")

plt.title("line chart")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend('best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot([100, 200, 300, 400, 500], [10, 20, 30, 40, 50], label="blue")

plt.title("line chart")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend('best')
plt.grid(color='black', linestyle='--', linewidth=0.1)
plt.show()
```

### xticks, yticks

```python
plt.figure(figsize=(10, 5))
plt.plot([100, 200, 300, 400, 500], [10, 20, 30, 40, 50], label="blue")

plt.title("line chart")
plt.xlabel('x label')
plt.ylabel('y label')

plt.xticks([i for i in range(80, 540, 20)])
plt.yticks([i for i in range(0, 70, 10)])

plt.legend('best')
plt.grid()
plt.show()
```

### axis

```python
plt.figure(figsize=(10, 5))
plt.plot([100, 200, 300, 400, 500], [10, 20, 30, 40, 50], label="blue")

plt.title("line chart")
plt.xlabel('x label')
plt.ylabel('y label')

plt.xticks([i for i in range(80, 540, 20)])
plt.yticks([i for i in range(0, 70, 10)])

plt.axis(False)
plt.legend()
plt.grid()
plt.show()
```

## 다양한 차트

```python
x = ["A", "B", "C", "D", "E"]
y = np.array([10, 20, 30, 40, 50])
plt.plot(x, y)
plt.show()
```

```python
x = ["A", "B", "C", "D", "E"]
y = np.array([10, 20, 30, 40, 50])
plt.bar(x, y)
plt.show()
```

```python
x = ["A", "B", "C", "D", "E"]
y = np.array([10, 20, 30, 40, 50])
plt.scatter(x, y)
plt.show()
```

```python
x = ["A", "B", "C", "D", "E"]
y = np.array([10, 20, 30, 40, 50])

plt.figure(figsize=(10, 5))

plt.subplot(131)
plt.plot(x, y, label="a")
plt.legend()

plt.subplot(132)
plt.bar(x, y)
plt.grid()

plt.subplot(133)
plt.scatter(x, y)
plt.axis(False)

plt.show()
```

```python
x = ["A", "B", "C", "D", "E"]
y = np.array([10, 20, 30, 40, 50])

plt.figure(figsize=(5, 10))

plt.subplot(311)
plt.plot(x, y)

plt.subplot(312)
plt.bar(x, y)

plt.subplot(313)
plt.scatter(x, y)

plt.show()
```
