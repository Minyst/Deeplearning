```python
change = {
    0:'A',
    1:'B',
    2:'C',
    3:'D',
    4:'E',
    5:'F',
    6:'G'
}
```

```python
target = train['대출등급']
tonum = {v:k for k,v in change.items()}
target = target.apply(lambda x: tonum[x])
print(target)
```

```python
train = train.drop(columns='ID',axis=1)
num_train = train.select_dtypes('number')
cat_train = train.select_dtypes('object')
num_test = test.select_dtypes('number')
cat_test = test.select_dtypes('object')
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
train_scaled = ss.fit_transform(num_train)
test_scaled = ss.transform(num_test)
train_scaled_df = pd.DataFrame(train_scaled,index=num_train.index,columns=num_train.columns)
test_scaled_df = pd.DataFrame(test_scaled,index=num_test.index,columns=num_test.columns)
train_scaled_df
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
ohe = OneHotEncoder()
train_encoded = ohe.fit_transform(cat_train)
train_encoded = train_encoded.toarray()
train_encoded = pd.DataFrame(train_encoded)
train_encoded.columns = ohe.get_feature_names_out()
train_encoded
```

```python
test_encoded = ohe.fit_transform(cat_test)
test_encoded = test_encoded.toarray()
test_encoded = pd.DataFrame(test_encoded)
test_encoded.columns = ohe.get_feature_names_out()
```

```python
final_train = pd.concat([train_scaled_df,train_encoded],axis=1)
final_test = pd.concat([test_scaled_df, test_encoded],axis=1)
```

```python
#KFold를 이용해서 검증만을 하는게 아니라 KFold로 만든 모델로 예측을 하고 싶은데 보통은 검증할때 x_valid로 예측을하는데 
#최종목적지는 final_test로 예측을 해야하기 때문에 이 부분을 어떻게 해결해야 하는지 궁금합니다.

from sklearn.model_selection import KFold
kf = KFold(n_splits=5,shuffle=True,random_state=1000)
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
lst = []
for train_index, val_index in kf.split(final_train):
    x_train = final_train.iloc[train_index]
    x_valid = final_train.iloc[val_index]
    y_train = target.iloc[train_index]
    y_valid = target.iloc[val_index]
    rf.fit(x_train,y_train)
    result = rf.predict(x_valid)
    lst.append(result)
avg_pred = lst.sum()/len(lst)
    
```
