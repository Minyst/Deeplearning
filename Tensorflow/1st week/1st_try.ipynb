```python
change = {
    0:'A',
    1:'B',
    2:'C',
    3:'D',
    4:'E',
    5:'F',
    6:'G'
}
```

```python
target = train['대출등급']
tonum = {v:k for k,v in change.items()}
target = target.apply(lambda x: tonum[x])
print(target)
```

```python
alldata = pd.concat([train,test])
alldata
```

```python
alldata = alldata.drop(columns='ID',axis=1)
alldata
```

```python
num = alldata.select_dtypes('number')
cat = alldata.select_dtypes('object')
```

```python
print(num)
```

```python
print(cat)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
num_scaled = ss.fit_transform(num)
num_df = pd.DataFrame(num_scaled,index=num.index,columns=num.columns)
num_df = num_df.reset_index()
num_df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
encoded_cat = ohe.fit_transform(cat)
encoded_cat_arr = encoded_cat.toarray()
encoded_cat_df = pd.DataFrame(encoded_cat_arr)
encoded_cat_df.columns = ohe.get_feature_names_out()
encoded_cat_df
```

```python
final = pd.concat([num_df,encoded_cat_df],axis=1)
train2 = final[:len(train)]
test2 = final[len(train):]
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from keras.callbacks import EarlyStopping
from tensorflow import keras
```

```python
k = len(change)
drop_rate = 0.5
model = Sequential()
model.add(keras.layers.Dense(1024, activation = 'relu'))
model.add(keras.layers.Dropout(drop_rate))
model.add(keras.layers.Dense(512, activation='relu'))
model.add(keras.layers.Dropout(drop_rate))
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dropout(drop_rate))
model.add(keras.layers.Dense(k, activation = 'softmax'))
model.compile(loss = "sparse_categorical_crossentropy", metrics = 'accuracy')
model.fit(train2, target, epochs = 5)
pred = model.predict(test2)
pred
```

```python
import numpy as np
final_pred = np.argmax(pred,1)
final_pred = [change[i] for i in final_pred]
final_pred
```

```python
samp['대출등급'] = final_pred
samp.to_csv('2nd_try.csv',index=False)
l = pd.read_csv('2nd_try.csv')
print(l)
```
